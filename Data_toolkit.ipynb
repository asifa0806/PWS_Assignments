**Q3- What is a Pandas DataFrame**

**Q12- What is the purpose of the describe() function in Pandas**

ans- The describe() function in Pandas is used to generate summary statistics of a DataFrame or a Series. It provides a quick overview of the distribution and key statistical metrics for numeric data, such as the mean, standard deviation, min, max, and percentiles.

Key Statistics Produced by describe():
Count: The number of non-null values in each column.
Mean: The average of the values in each column.
Standard deviation (std): Measures the spread or dispersion of the data.
Min: The minimum value in each column.
25th percentile (25%): The value below which 25% of the data falls.
50th percentile (50%): The median value of the data.
75th percentile (75%): The value below which 75% of the data falls.
Max: The maximum value in each column.

Values: The column(s) to aggregate.
Index: The column(s) to group by, forming the rows of the pivot table.
Columns: The column(s) to use as the columns of the pivot table (optional).
Aggregation Function: The function (e.g., sum, mean, count) used to aggregate the data.

Efficient Summary: They allow you to quickly summarize and aggregate large datasets in a compact format.

Better Data Insights: Pivot tables help to uncover trends, patterns, and relationships between different categories in your data.

Flexibility: You can easily manipulate, filter, and customize pivot tables for specific analyses.

```python

#Q4-How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset (using Seaborn's built-in 'iris' dataset)
df = sns.load_dataset('iris')

# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number']) # This line selects columns with numeric data types

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Customize the plot
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()

```

```python

#Q8-Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt

# Data for the lines
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # First line (y = x^2)
y2 = [1, 2, 3, 4, 5]    # Second line (y = x)
y3 = [25, 20, 15, 10, 5]  # Third line (y = 25 - x)

# Create a line plot with multiple lines
plt.plot(x, y1, label="y = x^2", color='blue', marker='o')  # First line
plt.plot(x, y2, label="y = x", color='green', linestyle='--')  # Second line
plt.plot(x, y3, label="y = 25 - x", color='red', linestyle='-.')

# Add labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Multiple Line Plot Example")

# Show legend
plt.legend()

# Display the plot
plt.show()

```
