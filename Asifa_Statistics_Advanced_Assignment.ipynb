```python
#Question3- Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The value (0 or 1) for which to calculate the probability.
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability P(X = x).
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

# Example usage
p = 0.7  # Probability of success
print(bernoulli_pdf(1, p))  # Output: 0.7
print(bernoulli_pdf(0, p))  # Output: 0.3
```

```python
#Question6-Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

def plot_cdf(a, b):
    """Plots the CDF of a discrete uniform distribution."""
    x = np.arange(a, b+2)  # Include b+1 for step function behavior
    cdf = randint.cdf(x, a, b+1)

    plt.step(x, cdf, where='post', marker='o', label=f'Discrete Uniform({a}, {b})')
    plt.xlabel('x')
    plt.ylabel('CDF')
    plt.title('Cumulative Distribution Function (CDF)')
    plt.xticks(np.arange(a, b+1, 1))
    plt.yticks(np.linspace(0, 1, b-a+2))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# Example usage
a, b = 1, 6  # A fair die (1 to 6)
plot_cdf(a, b)
```

```python
#Question9- Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Convert the input data to a numpy array for easier manipulation
    data = np.array(data)

    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plotting the Z-scores
    plt.figure(figsize=(10, 6))
    plt.plot(z_scores, marker='o', linestyle='-', color='b', label='Z-scores')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title("Z-scores of the Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-score")
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
data = [10, 12, 23, 23, 16, 23, 21, 25, 27, 30, 35, 42, 10]
z_scores = calculate_and_plot_z_scores(data)
```

```python
#Question18- Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import scipy.stats as stats

def calculate_z_score(data, x):
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-score for the given data point x
    z_score = (x - mean) / std_dev

    return z_score

def compare_to_standard_normal_distribution(z_score):
    # We can use the CDF (cumulative distribution function) of the standard normal distribution
    p_value = stats.norm.cdf(z_score)  # This gives the probability for this Z-score

    if z_score > 0:
        comparison = "above average"
    elif z_score < 0:
        comparison = "below average"
    else:
        comparison = "exactly average"

    return p_value, comparison

# Example usage:
data = [12, 15, 14, 10, 13, 16, 14, 17, 11, 15]  # Sample data
x = 18  # Data point to calculate Z-score for

z_score = calculate_z_score(data, x)
p_value, comparison = compare_to_standard_normal_distribution(z_score)

print(f"Data: {data}")
print(f"Given data point: {x}")
print(f"Z-score: {z_score:.2f}")
print(f"CDF value for this Z-score: {p_value:.4f}")
print(f"This data point is {comparison} compared to the average.")
```

```python
#Question19- Implement hypothesis testing using Z-statistics for a sample dataset.
import math
from scipy.stats import norm

# Function to perform Z-test
def z_test(sample_data, population_mean, population_std_dev, alpha=0.05):
    # Sample statistics
    sample_size = len(sample_data)
    sample_mean = sum(sample_data) / sample_size

    # Calculate Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std_dev / math.sqrt(sample_size))

    # Critical Z-value for two-tailed test (95% confidence level, alpha = 0.05)
    z_critical = norm.ppf(1 - alpha / 2)

    # Print Z-statistic and critical value
    print(f"Z-statistic: {z_statistic}")
    print(f"Critical Z-value (for alpha={alpha}): {z_critical}")

    # Determine if we reject the null hypothesis
    reject_null = abs(z_statistic) > z_critical
    if reject_null:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference from the population mean.")

# Example usage
if __name__ == "__main__":
    # Sample data
    sample_data = [105, 110, 98, 95, 106, 109, 104, 107, 111, 103, 102, 99, 108, 106, 100, 104, 107, 110, 103, 106, 101, 107, 105, 108, 102, 103, 106, 105, 104, 107]

    # Population parameters
    population_mean = 100  # This is the value from the null hypothesis
    population_std_dev = 15  # Assumed population standard deviation

    # Perform Z-test
    z_test(sample_data, population_mean, population_std_dev, alpha=0.05)
```

```python
#Question20- Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

# Sample dataset
data = [10, 12, 13, 15, 14, 16, 12, 11, 14, 13]

# Calculate sample mean and standard error
mean = np.mean(data)
std_error = np.std(data, ddof=1) / np.sqrt(len(data))

# Confidence level (95%)
confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate the t-statistic for the given confidence level and degrees of freedom
degrees_of_freedom = len(data) - 1
t_critical = stats.t.ppf(1 - alpha/2, degrees_of_freedom)

# Calculate the margin of error
margin_of_error = t_critical * std_error

# Calculate the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Output the results
print(f"Sample Mean: {mean}")
print(f"Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
#Question21- Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats

def generate_data(mean=50, std_dev=10, sample_size=100):
    """Generate random data from a normal distribution."""
    return np.random.normal(loc=mean, scale=std_dev, size=sample_size)

def calculate_confidence_interval(data, confidence=0.95):
    """Calculate the confidence interval for the mean of the data."""
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)

    # Compute the margin of error
    margin_error = stats.t.ppf((1 + confidence) / 2, df=n-1) * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error

    return sample_mean, (lower_bound, upper_bound)

def main():
    """Generate data, compute confidence interval, and interpret results."""
    data = generate_data()
    mean, conf_interval = calculate_confidence_interval(data)

    print(f"Sample Mean: {mean:.2f}")
    print(f"95% Confidence Interval: ({conf_interval[0]:.2f}, {conf_interval[1]:.2f})")
    print("Interpretation: We are 95% confident that the true population mean lies within this interval.")

if __name__ == "__main__":
    main()
```

```python
#Question22- Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std_dev=1, num_points=1000):
    # Generate x values
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, num_points)

    # Compute the PDF
    pdf_values = norm.pdf(x, mean, std_dev)

    # Plot the PDF
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf_values, label=f'N({mean}, {std_dev})', color='blue')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.title('Normal Distribution PDF')
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
plot_normal_pdf(mean=0, std_dev=1)
```

```python
#Question23- Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poison distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam, x_max=20):
    """
    Plots the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
    lam (float): The average rate (lambda) of the Poisson distribution.
    x_max (int): The maximum value of x to consider in the plot.
    """
    x = np.arange(0, x_max + 1)
    cdf_values = poisson.cdf(x, lam)

    plt.figure(figsize=(8, 5))
    plt.step(x, cdf_values, where='mid', label=f'Poisson CDF (={lam})', color='b')
    plt.xlabel('x')
    plt.ylabel('Cumulative Probability')
    plt.title('Poisson CDF')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    lambda_value = 5  # Set the lambda value for the Poisson distribution
    plot_poisson_cdf(lambda_value)
```

```python
#Question25- Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def generate_data(mean1=50, std1=10, size1=100, mean2=50, std2=20, size2=100):
    """Generate two datasets with specified mean, standard deviation, and size."""
    data1 = np.random.normal(mean1, std1, size1)
    data2 = np.random.normal(mean2, std2, size2)
    return data1, data2

def compute_std(data1, data2):
    """Compute and return the standard deviations of two datasets."""
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)
    return std1, std2

def visualize_data(data1, data2, std1, std2):
    """Visualize the data using histograms and boxplots."""
    plt.figure(figsize=(12, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data1, bins=20, kde=True, color='blue', label=f'Dataset 1 (std={std1:.2f})', alpha=0.6)
    sns.histplot(data2, bins=20, kde=True, color='red', label=f'Dataset 2 (std={std2:.2f})', alpha=0.6)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.title('Histogram Comparison')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(data=[data1, data2], palette=['blue', 'red'])
    plt.xticks([0, 1], [f'Dataset 1 (std={std1:.2f})', f'Dataset 2 (std={std2:.2f})'])
    plt.ylabel('Value')
    plt.title('Boxplot Comparison')

    plt.tight_layout()
    plt.show()

def main():
    """Main function to generate, compute, and visualize data."""
    data1, data2 = generate_data()
    std1, std2 = compute_std(data1, data2)
    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")
    visualize_data(data1, data2, std1, std2)

if __name__ == "__main__":
    main()
```

**Question1**- What is a random variable in probability theory?
**Answer**-In probability theory, a random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment. It essentially maps outcomes of a random process to real numbers.

Types of Random Variables:
1. Discrete Random Variable  Takes on a finite or countable number of possible values (e.g., number of heads in 3 coin tosses).

2. Continuous Random Variable  Takes on an infinite number of possible values within a given range (e.g., the height of a randomly chosen person).



**Question2**-What are the types of random variables?
**Answer**-Random variables are classified into two main types:

1. Discrete Random Variables

1) These take on a finite or countably infinite number of distinct values.

2) Examples: The number of heads in 10 coin flips, the number of customers arriving at a store in an hour.

3) Probability is assigned using a probability mass function (PMF).

2. Continuous Random Variables

1) These can take any value within a given range (uncountably infinite).

2) Examples: The height of students in a class, the time it takes to run a marathon.

3) Probability is assigned using a probability density function (PDF).



**Question3**- What is the difference between discrete and continuous distributions?
**Answer**- The main difference between discrete and continuous distributions lies in the type of values they take:

1. Discrete Distributions

- Deal with countable outcomes (e.g., whole numbers).

- Probabilities are assigned to individual values.

- Example: The number of heads when flipping a coin three times follows a binomial distribution.

2. Common Discrete Distributions:

- Binomial Distribution (number of successes in fixed trials)

- Poisson Distribution (number of events in a fixed interval)

- Geometric Distribution (number of trials until the first success)

3. Continuous Distributions

- Deal with uncountable values (e.g., real numbers).

- Probabilities are assigned over an interval using a probability density function (PDF).

- Example: The height of people follows a normal distribution.

4. Common Continuous Distributions:

- Normal Distribution (bell-shaped curve, used in many natural phenomena)

- Exponential Distribution (time between events in a Poisson process)

- Uniform Distribution (equal probability across an interval)
