```python
#Q-1-Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
#Ans
def sum_of_even_numbers(nums):
    return sum(num for num in nums if num % 2 == 0)
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)  # Output: 12 (2 + 4 + 6)
```

```python
#Q-4 Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    # Handle numbers less than 2 (not prime)
    if num < 2:
        return False
    # Check for divisibility from 2 to the square root of the number
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False  # Divisible by another number, so not prime
    return True
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")
```

```python
#Q-5- #Question5- Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#Answer

class FibonacciIterator:
    def __init__(self, n):
        # Initialize the number of terms and the first two Fibonacci numbers
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        # Return the iterator object itself
        return self

    def __next__(self):
        # If the number of terms to generate is reached, stop the iteration
        if self.count >= self.n:
            raise StopIteration

        # Generate the next Fibonacci number and update a and b
        fib_num = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib_num
# Create an iterator for the first 10 Fibonacci numbers
fib_iterator = FibonacciIterator(10)

# Iterate through the Fibonacci sequence and print the numbers
for num in fib_iterator:
    print(num)
```

```python
#Question6- Write a generator function in Python that yields the powers of 2 up to a given exponent.
#Answer
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
# Create a generator that yields powers of 2 up to 2^5
for power in powers_of_two(5):
    print(power)
```

```python
#Question7- Implement a generator function that reads a file line by line and yields each line as a string.
#Answer
def read_file_line_by_line(file_path):
    """
    Generator function to read a file line by line.

    :param file_path: Path to the file to be read.
    :yield: Each line in the file as a string.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
if __name__ == "__main__":
    file_path = r"/content/Example.txt"
    for line in read_file_line_by_line(file_path):
        print(line)
```

```python
#Question8- Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
#Answer
# List of tuples
tuples_list = [(1, 5), (2, 3), (4, 1), (3, 2)]

# Sort the list of tuples by the second element using a lambda function
sorted_list = sorted(tuples_list, key=lambda x: x[1])

print(sorted_list)
```

```python
#Question9-Write a Python program that uses map to convert a list of temperatures from Celsius to Fahrenheit.
#Answer

# List of temperatures in Celsius
celsius_temps = [0, 20, 30, 40, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map to apply the conversion to each temperature in the list
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```

```python
#Question10- Create a Python program that uses filter) to remove all the vowels from a given string.
#Answer
# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Input string
input_string = "Hello, World!"

# Use filter to remove vowels and join the result back into a string
filtered_string = ''.join(filter(is_not_vowel, input_string))

print(filtered_string)
```

```python
#Question11-Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this
#Answer
transactions = [
    ['Book A', 5, 10.0],  # 5 copies of Book A sold at $10 each
    ['Book B', 2, 15.0],  # 2 copies of Book B sold at $15 each
    ['Book C', 8, 7.5],   # 8 copies of Book C sold at $7.5 each
    ['Book D', 3, 12.0]   # 3 copies of Book D sold at $12 each
]
def calculate_total_revenue(transactions):
    total_revenue = 0
    for transaction in transactions:
        # Each transaction is a list: [book_title, quantity_sold, price_per_copy]
        quantity_sold = transaction[1]
        price_per_copy = transaction[2]

        # Revenue for this transaction
        revenue = quantity_sold * price_per_copy
        total_revenue += revenue

    return total_revenue

# Example usage
transactions = [
    ['Book A', 5, 10.0],
    ['Book B', 2, 15.0],
    ['Book C', 8, 7.5],
    ['Book D', 3, 12.0]
]

total_revenue = calculate_total_revenue(transactions)
print(f"Total revenue: ${total_revenue:.2f}")
```

**-6- Explain the concept of generators in Python and how they are defined.**

Answer

Generators in Python
A generator in Python is a special type of iterable that allows you to iterate over a potentially large sequence of data, but without storing the entire sequence in memory at once. This is achieved by using a lazy evaluation approach, where values are generated on-the-fly as needed, rather than all at once. This can lead to significant memory savings when working with large datasets or streams of data.

A generator is a function that uses the yield keyword to produce a sequence of values, one at a time, as they are requested. When the function is called, it doesn't execute its code immediately but returns a generator object instead, which can then be iterated over.
